Список вопросов по работе с Issues на GitHub
Общие вопросы

    1. Что такое Issues на GitHub и для чего они используются?
        
        Это инструмент управления задачами, который позволяет отслеживать баги, фичи, улучшения и другие запросы.

    2. Чем Issues отличаются от других инструментов управления задачами? 

        Issues интегрирован в GitHub и может быть привязан к коммитам.

    3. Какие основные компоненты (поля) есть у каждого Issue?
    
        Title, Description, Type, Priority, Assignee, Milestone.

Создание Issues

    4. Как создать новое Issue в репозитории?

        -> Issues -> New Issue -> Title, Description, Assignee, Milestone, Type -> Create Issues. 


    5. Какие данные рекомендуется указывать в описании Issue для лучшего понимания задачи?

        Следует сделать детальное описание бага или улучшения, при этом следует указать как воспроизвести баг или как видишь улучшение. Для задач следует указать конечную цель и критерии выполнения.

    6. Какие теги (labels) можно добавить к Issue? Какие из них стандартные?

        bug, question, enhancement, documentation, help wanted, wontfix, high-priority, frontend.


    7. Как прикрепить Assignees (ответственных) к Issue?

        правый side bar -> Assignees -> add name -> Submit.

Работа с Issues

    8. Как использовать Labels для классификации задач?

        правый side bar -> Labels -> add label. Labels выбираются в зависимости от задачи.

    9. Для чего нужен Milestone, и как связать его с Issue?

        правый side bar -> Milestone -> add milestone. Нужен для определения крайнего срока выполнения задачи.

    10. Как привязать Issue к пул-реквесту (Pull Request)?

        Comment: Closes (ссылка на Issue или номер Issue).

    11. Как добавить комментарий к существующему Issue?

        Зайти на страницу с Issue -> перейти вниз станицы -> нажать на кнопку add comment.

Закрытие и завершение Issues

    12. Как закрыть Issue вручную?

        Зайти на страницу с Issue -> нажать Close Issue. 

    13. Можно ли автоматически закрыть Issue с помощью сообщения в коммите или пул-реквесте? Как это сделать?

        Да, текст коммита:
        Title: ISSUE NAME (ссылка на Issue или номер Issue)
        Comment: Closes (ссылка на Issue или номер Issue).
        Текст PR:
        Title: ISSUE NAME (ссылка на Issue или номер Issue)
        Comment: Closes (ссылка на Issue или номер Issue).

    14. Как повторно открыть закрытое Issue, если работа ещё не завершена?

    -> раздел issues -> closed -> выбор issue -> нажимаем reopen issue.

Фильтрация и поиск

    15. Как найти все открытые или закрытые Issues в репозитории?

        Раздел issue в репозитории отображает открытые issue, далее нажимаем вкладку closes - тут отображаются уже закрытые.
        Ещё вариант написать в строке поиска is:issue state:open или is:issue state:closed

    16. Как использовать фильтры для поиска Issues по меткам, исполнителям или другим критериям?

        На странице в Issues есть панель фильтров над списком issues, где возможно отфильтровать по различным параметрам. 

    17. Как сортировать Issues по приоритету, дате создания или другим параметрам?

        Сортировка осуществляется через панель фильтров: 
        приоритет - milestones; 
        дата создания -  newest;
        другие фильтры в панели.

Интеграции и автоматизация

    18. Как настроить автоматические уведомления о новых или изменённых Issues?

        Зайти в Issue, найти на правом сайдбаре кнопку subscribe.
    
    19. Что такое Projects в контексте GitHub, и как связать их с Issues?

        GitHub Project - это система управления проектами включающая себя Issues, pull request и другие инструменты.
        Связать можно так: зайти в Issue, найти на правом сайдбаре вкладку Project.

    20. Какие сторонние инструменты можно использовать для автоматизации работы с Issues (например, боты, Webhooks)?

        Возможно настроить уведомления в TG при новых Issue, использовать CLI-утилиты, подключать Ai ботов - возможностей для автоматизации очень много.  

Коллаборация

    21. Как упомянуть другого пользователя в комментарии к Issue?

        @ник_пользователя.

    22. Как запросить дополнительные данные или уточнения у автора Issue?

        Открываем интересующий нас файл ,с помощью shift выделяем интересующую часть кода, нажимаем copy permalink, создаем issue, вставляем ссылку на код в описание, в комментарии к Issue задаём интересующий вопрос.

    23. Что делать, если Issue неактуально или его нужно объединить с другим?

        - В случае неактуальности ставится метка wontfix, после чего Issue закрывается. Причина закрытия указывается в комментариях.
        - В случе дублирования ставится метка duplicate, добавляется ссылка на действующий Issue, а в действующем Issue добавляется ссылка на дублирующийся.

Практические аспекты

    24. Как использовать шаблоны для создания Issues?

        При создание нового Issue появляется список шаблонов.

    25. Что такое Linked Issues, и как создать связь между задачами?

        Linked Issues — это связь между задачами в GitHub, показывающая их зависимость. Создать её можно через раздел Development в правом сайдбаре Issue.

    26. Какие метрики (например, время выполнения) можно отслеживать с помощью Issues?

        Время выполнения, тип задачи, приоритет, зависимости задач и другие метрики в соответствии с атрибутами.

    27. Какие best practices рекомендуются при работе с Issues в команде?

        Понятный заголовок.
        Достаточно подробное описание, чтоб отобразить суть проблемы. 
        Указание четкие критерии выполнения.
        Стандартизировать создание задач.
        Использовать метрики, milestone.
        Ссылаться на коммиты и PR.
        Описывать решение после выполнения задачи в комментариях.
